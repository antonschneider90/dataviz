```
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from numpy import log


```

```
from vega_datasets import data
unempl_data = data.unemployment_across_industries ()
unempl_data.head ()
```

```
year_check = []
for year in unempl_data.year:
  if year > 2005:
    year_check.append('Recent')
  else:
    year_check.append('Old')
```

```
unempl_data['year_chk'] = year_check
unempl_data.head()
```

```
unempl_data.series.unique()
```

```
plt.figure(figsize=(20, 10))
sns.lineplot(x='year', y='rate', hue = 'series', data=unempl_data)
plt.title('unemployment rates over years \n')
plt.legend(loc='upper left')
plt.show()
```

```
unempl_data_pivoted = unempl_data.pivot(columns='series', index='date', values='rate')
unempl_data_pivoted.head()
```
